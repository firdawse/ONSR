```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
import random

# Define the lists of values for each feature
D_cities = ['Fes','Fes','Fes','Fes','Fes','Fes','Fes','Fes','Fes', 'Casa', 'Rabat','El jadida','Fes','Fes','El jadida','Fes','Benguerir','Marrakech','Tanger','Kenitra','youssofia','khouribga']
A_cities = ['Fes', 'Casa', 'Rabat','El jadida','El jadida','El jadida','El jadida','El jadida','El jadida','El jadida','El jadida','Fes','Fes','El jadida','Fes','Benguerir','Marrakech','Tanger','Kenitra','yousofia','khouribga']
Travel_types = ['tgv 1st class','tgv 2nd class','Atlas 1st class','Atlas 2nd class','Atlas 2nd class','Atlas 2nd class','Atlas 2nd class','Atlas 2nd class','Atlas 2nd class','Atlas 2nd class','Atlas 2nd class','Atlas 2nd class','Atlas 2nd class']
prices = np.random.randint(31,400, size=10000)
A_time = ['6:00','8:00','8:00','8:00','8:00','8:00','8:00','8:00','8:00','10:00','12:00','13:00','14:00','15:00','16:00','17:00','18:00','19:00','20:00','21:00','22:00','23:00','00:00']
D_time = ['6:00','8:00','10:00','12:00','13:00','14:00','14:00','14:00','14:00','14:00','14:00','14:00','14:00','14:00','14:00','15:00','16:00','17:00','18:00','19:00','20:00','21:00','22:00','23:00','00:00']

targets = np.random.randint(0,1, size=10000)

# Create a dictionary of the data
data = {'D_cities': np.random.choice(D_cities, size=10000),
        'A_cities': np.random.choice(A_cities, size=10000),
        'D_time': np.random.choice(D_time, size=10000),
        'Travel_types':np.random.choice(Travel_types,size=10000),
        'A_time': np.random.choice(A_time, size=10000),
        'Price': prices,
        'target': targets}

# Create a dataframe from the dictionary
df = pd.DataFrame(data)
df['target'] = 0

df
```

```python
for i,d in df.iterrows():
    if (d['D_cities'] == 'Fes' or d['D_cities'] == 'El jadida' ) and (d['A_cities'] == 'Fes' or d['A_cities'] == 'El jadida') :
        df['target'][i] = 1

df
```

```python
A_hour = []
for time_str in df['A_time']:
    hour, minute = time_str.split(':')
    A_hour.append(int(hour))

df.drop('A_time', axis=1, inplace=True)
df['A_time'] = A_hour
```

```python
D_hour = []
for time_str in df['D_time']:
    hour, minute = time_str.split(':')
    D_hour.append(int(hour))

df.drop('D_time', axis=1, inplace=True)        
df['D_time'] = D_hour
```

```python
df
```

```python
features = df.columns
```

```python
feature = features.drop("target")
```

```python
X = df[feature]
```

```python
y = df['target']
```

```python
X
```

```python
# split the data into train and test sets
x_train, x_test, y_train, y_test= train_test_split(X, y,
                                                   test_size= 0.2,
                                                   shuffle= True, #shuffle the data to avoid bias
                                                   random_state= 0)
print(f'training set size: {x_train.shape[0]} samples \ntest set size: {x_test.shape[0]} samples')
```

```python
# create a knn, k = 100
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=100)
```

```python
# fit to the train set
knn.fit(x_train, y_train);
```

```python
ytrain_pred = knn.predict(x_train) # predict on train set
ytest_pred = knn.predict(x_test) # predict on test set
```

```python
# accuracy
from sklearn.metrics import accuracy_score
acc_train = accuracy_score(y_train,ytrain_pred)
acc_test =accuracy_score(y_test, ytest_pred)
print(f"Train set: {acc_train}")
print(f"Test set: {acc_test}")
```
