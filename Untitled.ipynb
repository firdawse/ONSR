```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('path.jpeg')
img=img[50:,50:]
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply binary thresholding
thresh_value = 200  # threshold value, can be adjusted
max_value = 255  # maximum value to use with thresholding
_, binary_img = cv2.threshold(gray, thresh_value, max_value, cv2.THRESH_BINARY)

# Apply a Gaussian blur to reduce noise
blur = cv2.GaussianBlur(binary_img, (5, 5), 0)

# Apply Canny edge detection to detect edges
edges = cv2.Canny(blur, 50, 150)

# Apply a Hough transform to detect lines
lines = cv2.HoughLinesP(edges, rho=1, theta=np.pi/180, threshold=50, minLineLength=50, maxLineGap=10)

# Find the center of the road
if lines is not None:
    x = []
    y = []
    for line in lines:
        x1, y1, x2, y2 = line[0]
        x.append(x1)
        x.append(x2)
        y.append(y1)
        y.append(y2)
    center_x = int(np.mean(x))
    center_y = int(np.mean(y))

    # Draw a circle on the center of the road
    cv2.circle(img, (center_x, center_y), 7, (0, 0, 255), -1)

# Display the result
# Display the image
plt.imshow(img)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('path.jpeg')

rects = [(0, 125, 300, 25)]

# Process each rectangle
for rect in rects:
    x, y, w, h = rect

    # Draw the rectangle
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)


    # Find the center of the road
    crop = img[y:y+h, x:x+w]
    gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
    thresh_value = 200  # threshold value, can be adjusted
    max_value = 255  # maximum value to use with thresholding
    _, binary_img = cv2.threshold(gray, thresh_value, max_value, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(binary_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) > 0:
        c = max(contours, key=cv2.contourArea)
        M = cv2.moments(c)
        if M["m00"] != 0:
            cx = int(M["m10"] / M["m00"])
            cy = int(M["m01"] / M["m00"])
            cv2.circle(crop, (cx, cy), 7, (0, 255, 0), -1)


# Display the image
plt.imshow(crop)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('path.jpeg')

# Define the three rectangles
rects = [(0, 115, 800, 25),(0, 140, 800, 25),(0, 165, 800, 25)]
X=[]
Y=[]
# Process each rectangle
for rect in rects:
    x, y, w, h = rect

    # Draw the rectangle
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Find the center of the road
    crop = img[y:y+h, x:x+w]
    gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
    thresh_value = 165  # threshold value, can be adjusted
    max_value = 255  # maximum value to use with thresholding

    _, binary_img = cv2.threshold(gray, thresh_value, max_value, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(binary_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) > 0:
        c = max(contours, key=cv2.contourArea)
        M = cv2.moments(c)
        if M["m00"] != 0:
            cx = int(M["m10"] / M["m00"])
            cy = int(M["m01"] / M["m00"])
            X.append(cx)
            Y.append(cy+y)
            cv2.circle(crop, (cx, cy), 7, (0, 255, 0), -1)

# Create a matrix of the form [x 1] for each point
A = np.vstack([Y, np.ones(len(Y))]).T

# Calculate the least-squares solution for the line coefficients
m, c = np.linalg.lstsq(A, X, rcond=None)[0]

# Print the slope and intercept of the line
print('Slope:', X)
print('Intercept:', Y)
print(np.arctan(m) * 180 / np.pi)
#Plot the points and the line
x1 = 250*m+c
x2 = 75*m+c
print(x1)
cv2.line(img,(int(x1),250),(int(x2),75),color = (255, 0, 0))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
